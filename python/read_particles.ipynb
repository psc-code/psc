```python
df1 = df.iloc[::100]
df1[df1.q==-1].plot.scatter('z', 'y', marker='.')
plt.gca().set_aspect(1);
#plt.xlim(4, 6)
#plt.ylim(4, 6)
plt.tight_layout()
```

```python
def plot_distribution(df1):
    fig, ax = plt.subplots(1, 2)
    df1[df1.q==-1].plot.hexbin('px', 'pz', gridsize=50, ax=ax[0]);
    ax[0].set_xlim(-.5, .5)
    ax[0].set_ylim(-.5, .5)
    ax[0].set_aspect(1.)
    df1[df1.q==1].plot.hexbin('px', 'pz', gridsize=50, ax=ax[1]);
    ax[1].set_xlim(-.5, .5)
    ax[1].set_ylim(-.5, .5)
    ax[1].set_aspect(1.)
```

```python
plot_distribution(df)
```
